```
# loop through these for each faction
soup = bs(allContent[0]) 
mydivs = soup.find_all("div", {"class": "datasheet"})

# loop through divs for each unit
firstTable = mydivs[0].findAll("table")[0] 


class unit:
  def __init__(self, element):
    self.element = element

    self.secondRow = self.element.findAll("tr")[1]
    self.price = self.secondRow.find_next("div").text
    self.baseSize = self.secondRow.find_next("span").text
    self.name = self.secondRow.text.split(self.price)[1].split(self.baseSize)[0]

    self.thirdRow = self.element.findAll("tr")[2]
    self.numberOfUnits = self.thirdRow.findAll("td")[0].text
    self.movement = self.thirdRow.findAll("td")[2].text
    self.ws = self.thirdRow.findAll("td")[3].text
    self.bs = self.thirdRow.findAll("td")[4].text
    self.s = self.thirdRow.findAll("td")[5].text
    self.t = self.thirdRow.findAll("td")[6].text
    self.w = self.thirdRow.findAll("td")[7].text
    self.a = self.thirdRow.findAll("td")[8].text
    self.ld = self.thirdRow.findAll("td")[9].text
    self.sv = self.thirdRow.findAll("td")[10].text



  def print(self):
    print("price: " + self.price)
    print("baseSize: " + self.baseSize)
    print("name: " + self.name)
    print("numberOfUnits: " + self.numberOfUnits)
    print("movement: " + self.movement)
    print("ws: " + self.ws)
    print("bs: " + self.bs)
    print("s: " + self.s)
    print("t: " + self.t)
    print("w: " + self.w)
    print("a: " + self.a)
    print("ld: " + self.ld)
    print("sv: " + self.sv)



firstUnit = unit(firstTable)
#firstUnit.print()







```
